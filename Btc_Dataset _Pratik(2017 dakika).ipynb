```python
import pandas as pd
```

# Bilgiyi alma, sutünları düşürme, datetime ayarlama, küçükten büyüğe zamana göre ayarlama, Nan değerleri doldurma

```python
df = pd.read_csv("btc.csv")
df=df.drop('unix',axis=1)
df=df.drop('symbol',axis=1)
df['date'] = pd.to_datetime(df['date'], errors='coerce')
df = df.sort_values(by="date",ascending=[True])
df.loc[:,df.columns != 'date'] = df.loc[:, df.columns != 'date'].fillna(method='ffill')
df = df.reset_index(drop=True)
columns_to_add = ['open', 'close', 'high', 'low', 'Volume BTC', 'Volume USD']
for col in columns_to_add:
    df['5m_'+col] = pd.NA  # Yeni sütunları ekliyoruz, başlangıçta boş
```

```python

df = df.loc[:20000]
    # 'open' sütununu işleyelim (dakika son değeri 1 veya 6)
for i in range(1, len(df)):
    current_minute_str = str(df['date'].dt.minute.loc[i])  # Dakika bilgisini al
    current_minute = int(current_minute_str[-1])
    if current_minute == 0:
        df.loc[i:i+4,'5m_open'] = df.loc[i,'open']
    elif current_minute == 5:
        df.loc[i:i+4,'5m_open'] = df.loc[i,'open']
```

### Herhangi bir durumda aşağıda sütun ekleme yedek kodları var.

```python
###Yedek kod bölümü

columns_to_add = ['open', 'close', 'high', 'low', 'Volume BTC', 'Volume USD']
for col in columns_to_add:
    df['5m_'+col] = pd.NA  # Yeni sütunları ekliyoruz, başlangıçta boş
    # 'open' sütununu işleyelim (dakika son değeri 1 veya 6)
for i in range(1, len(df)):
    current_minute_str = str(df['date'].dt.minute.loc[i])  # Dakika bilgisini al
    current_minute = int(current_minute_str[-1])
    if current_minute == 0:
        df.loc[i:i+4,'5m_open'] = df.loc[i,'open']
    elif current_minute == 5:
        df.loc[i:i+4,'5m_open'] = df.loc[i,'open']
  # 'close' sütununu işleyelim (dakika son değeri 4 veya 9)
for i in range(0, len(df)):
    current_minute_str = str(df['date'].dt.minute.loc[i])  # Dakika bilgisini al
    current_minute = int(current_minute_str[-1])
    if current_minute == 4:
        df.loc[i-4:i,'5m_close'] = df.loc[i,'close']
    elif current_minute == 9:
        df.loc[i-4:i,'5m_close'] = df.loc[i,'close']
  # 'high' ve 'low' sütunlarını işleyelim (4 veya 9)
for i in range(0, len(df)):
    current_minute_str = str(df['date'].dt.minute.loc[i])  # Dakika bilgisini al
    current_minute = int(current_minute_str[-1])
    if current_minute == 4:
        df.loc[i-4:i, '5m_high'] = df.loc[i-4:i, 'high'].max()
        df.loc[i-4:i+4, '5m_low'] = df.loc[i-4:i, 'low'].min()
    elif current_minute == 9:
        df.loc[i-4:i, '5m_high'] = df.loc[i-4:i, 'high'].max()
        df.loc[i-4:i, '5m_low'] = df.loc[i-4:i, 'low'].min()
 # 'Volume BTC' ve 'Volume USD' sütunlarını işleyelim (dakika 0-4 veya 5-9)
for i in range(0, len(df)):
    current_minute_str = str(df['date'].dt.minute.loc[i])  # Dakika bilgisini al
    current_minute = int(current_minute_str[-1])
    if current_minute == 4:
      df.loc[i-4:i, '5m_Volume BTC'] = df.loc[i-4:i, 'Volume BTC'].sum()
      df.loc[i-4:i, '5m_Volume USD'] = df.loc[i-4:i, 'Volume USD'].sum()
    elif current_minute == 9:
      df.loc[i-4:i, '5m_Volume BTC'] = df.loc[i-4:i, 'Volume BTC'].sum()
      df.loc[i-4:i, '5m_Volume USD'] = df.loc[i-4:i, 'Volume USD'].sum()
```
