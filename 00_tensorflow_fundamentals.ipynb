# Creating Random Tensors
Random Tensors are tensors of some arbitrary size which contain random numbers

```python
random_1 = tf.random.Generator.from_seed(42) # set seed for reproducibility
random_1 = random_1.normal(shape=(3,2))
random_2 = tf.random.Generator.from_seed(42)
random_2 = random_2.normal(shape=(3,2))
#EÅŸitler mi deÄŸiller mi yukarÄ±daki rastgele olan random tensÃ¶rler
random_1, random_2, random_1 == random_2
#EÅŸitler Ã§Ã¼nkÃ¼ seed (Ã§ekirdek) aynÄ±
```

```python
# Shuffle = karÄ±ÅŸtÄ±rma // Verileri karÄ±ÅŸtÄ±rÄ±r. SÃ¼rekli veriler ajanÄ±n Ã¶ÄŸrenmesini etkilemesin
not_shuffled = tf.constant([[10,7],
                           [3,4],
                           [2,5]])
# Shuffle
tf.random.shuffle(not_shuffled)
```

```python
# Shuffle with seed/ bÃ¶yle yapÄ±nca sÃ¼rekli random veriyor.
tf.random.shuffle(not_shuffled, seed = 42) # Operation level seed
# Global random seed tanÄ±mlarsan sÃ¼rekli aynÄ± shuffle tÃ¼rÃ¼ yapar.
tf.random.set_seed(42) # Global random seet
tf.random.shuffle(not_shuffled) # bÃ¶ylelikle sÃ¼rekli aynÄ± shuffle Ä± verir.
```

**Matrix multiplication**

In machine learning, matrix multiplication is one of the comman tensor operation

There are two rules our tensors(or metrices) need to fullfil if we're going to matrix multiply them:
 1. The inner dimensions must match
 2. The resulting matrix has the shape of the outer dimensions

```python
# Matrix multiplication in tensorflow
print(tensor)
tf.matmul(tensor, tensor)
```

```python
# Matrix multiplication with Python operator "@"
tensor@tensor
```

```python
# Create a tensor (3,2) tensor
X = tf.constant([[1,2],
                [3,4],
                [5,6]])
# Create another (3,2) tensor
Y = tf.constant([[7,8],
                 [9,10],
                 [11,12]])
X,Y
```

```python
# Try to matrix multiply tensors of same shape
X @ Y
```

```python
#Let's change the shape of Y
X.shape, tf.reshape(Y, shape=(2,3)).shape
```

```python
# Try to matrix multiply X by reshaped Y
X @ tf.reshape(Y, shape=(2,3))
```

```python
tf.matmul(X, tf.reshape(Y, shape=(2,3)))
```

```python
# Try to change the shape of X instead of Y
tf.matmul(tf.reshape(X, shape=(2,3)), Y)
```

```python
# Can do the same with transpose
X,tf.transpose(X), tf.reshape(X,shape=(2,3))
```

```python
# Try matrix multiplication with transpose rather than reshape
tf.matmul(tf.transpose(X),Y)
```

**The Dot product**

Matrix multiplication is also referred to as the dot product.

You can perform matrix multiplication using:
 * *tf.matmul()*
 * *tf.tensordot()*
 * *@*


```python
X, Y
```

```python
# Perform the dot product on X and Y (requires Z or Y to be
tf.tensordot(tf.transpose(X),Y, axes=1)
```

```python
# Perform matrix multiplication between X and Y (transposed)
tf.matmul(X, tf.transpose(Y))
```

```python
# Perform matrix multiplication between X and Y (reshaped)
tf.matmul(X, tf.reshape(Y, shape=(2,3)))
```

```python
# Check the values of Y, reshape Y and transposed Y
print("'Normal Y:")
print(Y,"\n" )
print( "Y reshaped to (2,3):")
print(tf.reshape(Y, shape=(2,3)), "\n")
print("Y transposed:")
print(tf.transpose(Y))
```

```python
tf.matmul(X, tf.transpose(Y))
```

Generally, when performing matrix multiplication on two tensors and one of the axes doesn't line up, you will transpose(rather than reshape) one of the tensors to get satisfy the matrix multiplication rules.

### Aggregating tensors

Aggregating tensors = condensing them from multiple values down to a smaller amount of values

```python
# Get the absolute values
D = tf.constant([-7, -10])
D
```

```python
# Get the absolute values
tf.abs(D)
```

Let's go through the folloowing forms of aggregation:
* Get the minimum
* Get the maximum
* Get the mean of a tensor
* Get the sum of a tensor

```python
# Create a random tensor with values between 0 and 100 of size 50
E = tf.constant(np.random.randint(0,100, size=50))
E
```

```python
tf.size(E), E.shape, E.ndim
```

```python
# Find the minimum
tf.reduce_min(E)
```

```python
# Find the maximum
tf.reduce_max(E)
```

```python
# Find the mean
tf.reduce_mean(E)
```

```python
# Find the sum
tf.reduce_sum(E)
```

```python
# Find the variance of our tensor
tf.reduce_variance(E) # Wont work
```

```python
# To Find the variance of our tensor, we need access to tensorflow_probability
import tensorflow_probability as tfp
tfp.stats.variance(E)
```

```python
# Find the variance of our E tensor
tf.math.reduce_variance(tf.cast(E, dtype=tf.float32))
```

```python
tf.math.reduce_std(E) # With a type error always go to the documentry and chech example types
```

```python
# Find the Standard deviation
tf.math.reduce_std(tf.cast(E, dtype=tf.float32))
```

###**Find the positional maximum and minimum**



```python
# Create a new tensor for finding positional minimum and maximum
tf.random.set_seed(42)
F = tf.random.uniform(shape=[50])
F
```

```python
# Find the positional maximum
tf.argmax(F)
```

```python
# Index on our largest value position
F[tf.argmax(F)]
```

```python
# Find the max value of F
tf.reduce_max(F)
```

```python
# Check for equeality
F[tf.argmax(F)] == tf.reduce_max(F)
```

```python
# Find the positional minimum
tf.argmin(F)
```

```python
# Find the minimum, using the positional minimum index
F[tf.argmin(F)]
```
