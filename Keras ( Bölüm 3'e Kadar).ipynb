## Eleman Temelli
Bu işlemler tensörün her elemanına ayrı ayrı uygulanır.

```python
def naive_add(x,y):
  assert len(x.shape)==2 # x ve ye 2 boyutlu numpy tensörü olması
  assert x.shape == y.shape
  x = x.copy()
  for i in range(x.shape[0]):
    for j in range(x.shape[1]):
      x[i,j]+=y[i,j]
  return x
```

```python
x = np.array([[7,14,21],
             [7,14,21],[7,14,21]])
y = np.array([[2,2,2],
             [3,3,3],[5,5,5]])

z = naive_add(x,y)
z
# Ancak numpy bu gibi deflere gerek kalmadan işlemleri yapar.
z = x+y #de yapar aynısını
z
```

```python
def naive_relu(x):
  assert len(x.shape)==2 # x ve ye 2 boyutlu numpy tensörü olması
  x = x.copy()
  for i in range(x.shape[0]):
    for j in range(x.shape[1]):
      x[i,j]=max(x[i,j],0)
  return x
```

```python
x = np.array([[7,14,-21],
             [7,-14,21],[7,14,21]])
z = naive_relu(x)
z
# relu işlemi max değerleri çeker. yani eksi olanları sıfıra çeker.
z= np.maximum (x,0) # buda aknı şeyi yapamar
z
```

##Yayma Operasyonu

```python
def naive_add_matrix_and_vector(x,y):
    assert len(x.shape)==2
    assert len(y.shape)==1
    assert x.shape[1]==y.shape[0]

    x = x.copy()

    for i in range(x.shape[0]):
      for j in range(x.shape[1]):
        x[i,j] +=y[j]
    return x
```

```python
x = np.array([[7,14,21],
             [7,14,21],[7,14,21]])
y = np.array([1,2,3])

z = naive_add_matrix_and_vector(x,y)
z
# matris ve vektörü topladı ve matris verdi.
a = np.random.random((64,3,32,10))
b = np.random.random((32,10))
c = np.maximum(a,b)
c.shape
# bu da numpy kütüphanesinin bize sağladığı kolaylık.
```
