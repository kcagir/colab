# Importlar

```python
import pandas as pd
import yfinance as yf
from google.colab import files
from google.colab import drive
import shutil
import os
```

## Dosy

```python
prices_raw = pd.read_csv("/content/drive/MyDrive/btc1m-2013--2025-03-23.csv")
prices_raw.set_index('Timestamp', inplace=True)
```

```python
prices_raw
```

### "2025-02-13"- "2025-02-16" arası veriler eksik. çekmenin yolunu bulamadım henüz  

```python
dfr = yf.download(tickers="BTC-USD", start="2025-03-20", end="2025-03-24", period="1d", interval="1m")
dfr.columns = ['Open', 'High', 'Low', 'Close', 'Volume']
desired_order = ["Open", "Close",	"High",	"Low", "Volume"]
dfr = dfr[desired_order]
dfr.index.rename("Timestamp", inplace=True)
df1 = dfr
```

```python
df1
```

## Concat

```python
df_combined = pd.concat([prices_raw, df1], ignore_index=False)
```

```python
df_combined
```

# Dowload


```python
filename = "btc1m-2013--2025-03-23.csv"
df_combined.to_csv(filename, index=True)
```

```python
# Dosya yüklemek istediğiniz yolu belirleyin
source_path = f'/content/{filename}'  # Yüklemek istediğiniz dosyanın yolu
destination_path = '/content/drive/MyDrive'  # Drive'da kaydedilecek yer

# Dosyayı taşıma
shutil.copy(source_path, destination_path)
```
