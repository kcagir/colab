Silver Bullet İşlem Adımları
📌 1. Doğru Zaman Dilimini Belirle:

Londra Seansı: 10:00 – 11:00 UTC

New York Seansı: 15:00 – 16:00 UTC

Asya Seansı: 02:00 – 03:00 UTC

En çok tercih edilen zaman: New York 10:00–11:00 (Türkiye saatiyle 17:00–18:00)

📌 2. Önceki Likidite Bölgelerini Bul:

Günün en düşük ve en yüksek fiyatlarını işaretle.

Son birkaç saatteki önemli destek-direnç seviyelerini belirle.

Likidite havuzlarını (Equal Highs / Equal Lows) tespit et.

📌 3. Likidite Alımı (Stop Hunt) Bekle:

Fiyat, belirlediğin likidite seviyelerini "avlamak" için önce bir yön hareketi yapacaktır.

Eğer yukarıdaki likiditeyi aldıysa, düşüş beklenir.

Eğer aşağıdaki likiditeyi aldıysa, yükseliş beklenir.

📌 4. Fair Value Gap (FVG) Oluşumunu Kontrol Et:

Likidite alındıktan sonra güçlü bir mum hareketi olur ve bir FVG boşluğu bırakır.

Bu boşluğa geri çekilme olursa, işlem fırsatı doğar.

📌 5. Giriş Yap ve Çıkış Noktalarını Belirle:

Giriş: FVG bölgesine retest yapıldığında.

Stop Loss: En son likidite avının hemen üstü veya altı.

Hedef: 20-30 pip veya belirlenen bir sonraki likidite havuzu.

# %0,5 Değişim---Eklendi

```python
def check_price_movement(df, threshold=0.005, lookahead=5):
    signals = [0] * len(df)  # Initialize signal list with 0s

    for i in range(len(df) - lookahead):
        future_prices = df['Close'][i+1:i+1+lookahead]
        current_price = df['Close'][i]
        up_condition = any(future_prices >= current_price * (1 + threshold))
        down_condition = any(future_prices <= current_price * (1 - threshold))

        if up_condition and not down_condition:
            signals[i] = 1
        elif down_condition and not up_condition:
            signals[i] = -1
        else:
            signals[i] = 0

    df['Change'] = signals
    return df
```

## Fonksiyonlar

### High/Low

```python
def find_high_low(data, window):
    """Belirli bir pencere (örneğin 20, 40, 60 mum) içindeki en yüksek ve en düşük seviyeleri bulur."""
    data[f'Highest_high_{window}'] = data['High'].rolling(window).max()
    data[f'Lowest_low_{window}'] = data['Low'].rolling(window).min()
```

## High / Low Atama

### 4 Saatlik

```python
# 20, 40 ve 60 mumluk pencereleri uygula
for window in [4800, 9600, 14400]:  # Örnek küçük pencere boyutları, gerçek veriyle 20-40-60 kullanılır
    find_high_low(data, window)
    data.rename(columns={f'Highest_high_{window}':f'Highest_high_4h_{int(window/4/60)}'},inplace=True)
    data.rename(columns={f'Lowest_low_{window}':f'Lowest_low_4h_{int(window/4/60)}'},inplace=True)
```

### Saatlik

```python
# 20, 40 ve 60 mumluk pencereleri uygula
for window in [1200, 2400, 3600]:  # Örnek küçük pencere boyutları, gerçek veriyle 20-40-60 kullanılır
    find_high_low(data, window)
    data.rename(columns={f'Highest_high_{window}':f'Highest_high_1h_{int(window/60)}'},inplace=True)
    data.rename(columns={f'Lowest_low_{window}':f'Lowest_low_1h_{int(window/60)}'},inplace=True)
```

### 5m

```python
# 20, 40 ve 60 mumluk pencereleri uygula
for window in [100, 200, 300]:  # Örnek küçük pencere boyutları, gerçek veriyle 20-40-60 kullanılır
    find_high_low(data, window)
    data.rename(columns={f'Highest_high_{window}':f'Highest_high_5m_{int(window/5)}'},inplace=True)
    data.rename(columns={f'Lowest_low_{window}':f'Lowest_low_5m_{int(window/5)}'},inplace=True)
```

## **Görünüşe göre değerlerde hata var çünkü hepsi false görünüyor. Tekrar bakmam lazım.**

```python
# Belirli zaman dilimlerindeki en yüksek ve en düşük seviyeleri kontrol etme
data['is_bullish_4h'] = data['Close'] > data['Highest_high_4h_20']
data['is_bearish_4h'] = data['Close'] < data['Lowest_low_4h_20']

data['is_bullish_1h'] = data['Close'] > data['Highest_high_1h_20']
data['is_bearish_1h'] = data['Close'] < data['Lowest_low_1h_20']

data['is_bullish_5m'] = data['Close'] > data['Highest_high_5m_20']
data['is_bearish_5m'] = data['Close'] < data['Lowest_low_5m_20']

# Market Structure Break (MSB) Kontrolü
data['bullish_msb'] = data['Close'].shift(1) > data['Highest_high_1m_20']  # Önceki en yüksek kırıldı mı?
data['bearish_msb'] = data['Close'].shift(1) < data['Lowest_low_1m_20']    # Önceki en düşük kırıldı mı?

# Fair Value Gap (FVG) Tespiti
data['fvg_up'] = data['Low'].shift(1) > data['High'].shift(2)
data['fvg_down'] = data['High'].shift(1) < data['Low'].shift(2)
```
