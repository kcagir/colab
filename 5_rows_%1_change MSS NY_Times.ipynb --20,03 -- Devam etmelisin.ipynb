# %0,5 Deƒüi≈üim---Eklendi

```python
def check_price_movement(df, threshold=0.005, lookahead=5):
    signals = [0] * len(df)  # Initialize signal list with 0s

    for i in range(len(df) - lookahead):
        future_prices = df['Close'][i+1:i+1+lookahead]
        current_price = df['Close'][i]
        up_condition = any(future_prices >= current_price * (1 + threshold))
        down_condition = any(future_prices <= current_price * (1 - threshold))

        if up_condition and not down_condition:
            signals[i] = 1
        elif down_condition and not up_condition:
            signals[i] = -1
        else:
            signals[i] = 0

    df['Change'] = signals
    return df
```

## Fonksiyonlar

### High/Low

```python
def find_high_low(data, window):
    """Belirli bir pencere (√∂rneƒüin 20, 40, 60 mum) i√ßindeki en y√ºksek ve en d√º≈ü√ºk seviyeleri bulur."""
    data[f'Highest_high_{window}'] = data['High'].rolling(window).max()
    data[f'Lowest_low_{window}'] = data['Low'].rolling(window).min()
```

## High / Low Atama

### 4 Saatlik

```python
# 20, 40 ve 60 mumluk pencereleri uygula
for window in [4800, 9600, 14400]:  # √ñrnek k√º√ß√ºk pencere boyutlarƒ±, ger√ßek veriyle 20-40-60 kullanƒ±lƒ±r
    find_high_low(data, window)
    data.rename(columns={f'Highest_high_{window}':f'Highest_high_4h_{int(window/4/60)}'},inplace=True)
    data.rename(columns={f'Lowest_low_{window}':f'Lowest_low_4h_{int(window/4/60)}'},inplace=True)
```

### Saatlik

```python
# 20, 40 ve 60 mumluk pencereleri uygula
for window in [1200, 2400, 3600]:  # √ñrnek k√º√ß√ºk pencere boyutlarƒ±, ger√ßek veriyle 20-40-60 kullanƒ±lƒ±r
    find_high_low(data, window)
    data.rename(columns={f'Highest_high_{window}':f'Highest_high_1h_{int(window/60)}'},inplace=True)
    data.rename(columns={f'Lowest_low_{window}':f'Lowest_low_1h_{int(window/60)}'},inplace=True)
```

### 5m

```python
# 20, 40 ve 60 mumluk pencereleri uygula
for window in [100, 200, 300]:  # √ñrnek k√º√ß√ºk pencere boyutlarƒ±, ger√ßek veriyle 20-40-60 kullanƒ±lƒ±r
    find_high_low(data, window)
    data.rename(columns={f'Highest_high_{window}':f'Highest_high_5m_{int(window/5)}'},inplace=True)
    data.rename(columns={f'Lowest_low_{window}':f'Lowest_low_5m_{int(window/5)}'},inplace=True)
```

### 1m

```python
# 20, 40 ve 60 mumluk pencereleri uygula
for window in [20, 40, 60]:  # √ñrnek k√º√ß√ºk pencere boyutlarƒ±, ger√ßek veriyle 20-40-60 kullanƒ±lƒ±r
    find_high_low(data, window)
    data.rename(columns={f'Highest_high_{window}':f'Highest_high_1m_{int(window)}'},inplace=True)
    data.rename(columns={f'Lowest_low_{window}':f'Lowest_low_1m_{int(window)}'},inplace=True)
```

## is_Bullish or Bearish & MSB (Market Structure Break) -- Eklendi
20 - 40 - 60 pencereden bakmak gerekiyor.
Veri dakikalƒ±k olduƒüu i√ßin b√ºy√ºk zaman dilimlerini saate √ßevirmek i√ßin *60 ve ayrƒ±ca *20/40/60 ekledim.

4 saat = 4800 / 9600 / 14400

1 saat = 1200 / 2400 / 3600

5 dakika = 100 / 200 / 300

1 dakika = 20 / 40 / 60

## **G√∂r√ºn√º≈üe g√∂re deƒüerlerde hata var √ß√ºnk√º hepsi false g√∂r√ºn√ºyor. Tekrar bakmam lazƒ±m.**

```python
# Belirli zaman dilimlerindeki en y√ºksek ve en d√º≈ü√ºk seviyeleri kontrol etme
data['is_bullish_4h'] = data['Close'] > data['Highest_high_4h_20']
data['is_bearish_4h'] = data['Close'] < data['Lowest_low_4h_20']

data['is_bullish_1h'] = data['Close'] > data['Highest_high_1h_20']
data['is_bearish_1h'] = data['Close'] < data['Lowest_low_1h_20']

data['is_bullish_5m'] = data['Close'] > data['Highest_high_5m_20']
data['is_bearish_5m'] = data['Close'] < data['Lowest_low_5m_20']

# Market Structure Break (MSB) Kontrol√º
data['bullish_msb'] = data['Close'].shift(1) > data['Highest_high_1m_20']  # √ñnceki en y√ºksek kƒ±rƒ±ldƒ± mƒ±?
data['bearish_msb'] = data['Close'].shift(1) < data['Lowest_low_1m_20']    # √ñnceki en d√º≈ü√ºk kƒ±rƒ±ldƒ± mƒ±?

# Fair Value Gap (FVG) Tespiti
data['fvg_up'] = data['Low'].shift(1) > data['High'].shift(2)
data['fvg_down'] = data['High'].shift(1) < data['Low'].shift(2)
```
